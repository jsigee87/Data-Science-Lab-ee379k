#### Part A:

##### Discrete dependent but uncorrelated distribution

Consider a random variable x = [1,0,1] with probabilities $\frac{1}{4}, \frac{1}{2},\frac{1}{4}$ respectively.

Consider another random variable y = -1 if x = 0 and y =1 otherwise.

We can clearly see that x and y are dependent on each other and their covariance is zero since expected values of x and y are zero.
